5.Sort following NumPy array: 1: Sort array by the i th row 2: Sort the array by the jth column

```python
import numpy

print("Printing Original array")
sampleArray = numpy.array([[1,3,6],[9,8,10],[2,5,9]]) 
print (sampleArray)

print("Enter by which Row to sort (start:0)")
i =input()
i=int(i)

sortArrayByRow = sampleArray[:,sampleArray[i,:].argsort()]
print("Sorting Original array by", i ,"th row")
print(sortArrayByRow)

print("Enter by which Coloum to sort (start:0)")
j=int(input())

print("Sorting Original array by",j,"th column")
sortArrayByColumn = sampleArray[sampleArray[:,j].argsort()]
print(sortArrayByColumn)
```

8.Find the positions of:
- elements in x where its value is more than its corresponding element in y, and
- elements in x where its value is equals to its corresponding element in y.

```python
import numpy as np

def largestInColumn(mat, rows, cols):
    r=0
    c=0
    for i in range(cols):
        maxm = mat[0][i]
        for j in range(rows):
            if mat[j][i] > maxm:
                maxm = mat[j][i]
                r=i
                c=j
        print("x position ",r," ","y Postion ",c," ",maxm)

n = int(input("Enter number of row: "))
m=int(input("Enter number of column: "))
mat = np.empty([n,m], dtype = np.uint16) 
print("Array",mat)
largestInColumn(mat, n, m);

def equal(mat, rows, cols):
    r=0
    c=0
    for i in range(cols):
        eq = mat[0][i]
        for j in range(rows):
            if mat[j][i] ==eq:
                eq = mat[j][i]
                r=j
                c=i
        print("x position ",r," ","y Postion ",c," ","equal ",eq)
mat = [[1,2,3],[4,5,6],[1,2,3]]
print("Array",mat)
equal(mat, 3, 3);
```

10.Write a program to multiply two matrices of size (100, 100) in two methods:

(a) by using np . dot (mat_l, mat_2 )

(b) by using for-loops. Comapre the time of execution in both the cases. Check out the documentation of np . dot in case that is not familiar to you. 

```python
import numpy as np
import random

mat_1=100*np.random.rand(100,100)
mat_2=100*np.random.rand(100,100)
print("Matrix M1: \n",mat_1)
print("Matrix M2: \n",mat_2)
Mul_1=np.dot(mat_1,mat_2)
print("\n\n Multiplication using np.dot(): ",Mul_1)

Mul_2=np.zeros_like(mat_1)
for i in range(0,99):
  for j in range(0,99):
    for k in range(0,99):
      Mul_2[i,j]+=mat_1[i,k]*mat_2[k,j]

print("\n\n Multiplication using for loop: ",Mul_2)
```
